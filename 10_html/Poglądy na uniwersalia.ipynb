```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dane = {'Nominalizm': 11,
 'Konceptualizm': 9,
 'Umiarkowany realizm pojÄ™ciowy': 12,
 'Idealizm': 1,
 'Inne': 3}
```

```python
cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0.1, 0.65, 5)]

plt.pie(x=dane.values(),colors=colors, autopct='%1.1f%%')
plt.legend(labels=dane.keys(),loc='upper right',bbox_to_anchor=(1.65, 1))
plt.tight_layout()
plt.show()
```
